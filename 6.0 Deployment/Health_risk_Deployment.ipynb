```python
myDataset = pd.read_csv("Health_Risk_Prediction.csv")
df1 = myDataset.copy() 
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# ----------------print("Loaded successfully!\n")------------------------
# 1. Load the saved model and encoders
# ----------------------------------------
load_model = pickle.load(open("DT_Class_FinalModel.sav",'rb'))  #Read Binary from the file  

#with open("encoders.sav", "rb") as f:
 #   encoders = pickle.load(f)
#print("Loaded successfully!\n")

# Columns used for prediction
input_order = ['age', 'sleep','bmi','exercise','sugar_intake','smoking','alcohol']

label_columns = ['exercise','sugar_intake','smoking','alcohol','health_risk']
#label_columns = ['exercise','sugar_intake','smoking','alcohol']  #Removed the output column in input
encoders = {}

for col in label_columns:
    le = LabelEncoder()
    df1[col] = le.fit_transform(df1[col])   # <--- FIT happens here
    encoders[col] = le

pickle.dump(encoders, open("encoders.sav", "wb"))
print("File SAVED successfully!\n")
```

```python

# Columns used during training — MUST match exact order
input_order = [
    'age',
    'sleep',
    'bmi',
    'exercise',
    'sugar_intake',
 #   'profession',
    'smoking',
    'alcohol',
  #  'married'
]

# All label-encoded columns
label_columns = [
    'exercise',
    'sugar_intake',
   # 'profession',
    'smoking',
    'alcohol',
    #'married',
 #   'health_risk'  #This is output column
]


# ----------------------------------------------------------
# 2. Get user input
# ----------------------------------------------------------

def get_user_input():

    print("Enter values for prediction:\n")

    data = {
        "age": float(input("Age: ")),
        "sleep": float(input("Sleep: ")),
        "bmi": float(input("BMI: ")),
        "exercise": input("Exercise (low/medium/high): "),
        "sugar_intake": input("Sugar Intake (low/medium/high): "),
   #     "profession": input("Profession: "),
        "smoking": input("Smoking (yes/no): "),
        "alcohol": input("Alcohol (yes/no): "),
    #    "married": input("Married (yes/no): ")
    }

    return data


# ----------------------------------------------------------
# 3. Encode categorical input
# ----------------------------------------------------------

def encode_input(data):
    print("Encode the input values")
    encoded_values = []

    for col in input_order:
        value = data[col]
     #   print("inside for",value)
        # If column was label-encoded → use encoder
        if col in encoders:
            encoded_value = encoders[col].transform([value])[0]
        else:
            # numeric values
            encoded_value = float(value)

        encoded_values.append(encoded_value)

    return np.array([encoded_values])

# ----------------------------------------------------------
# 4. Decode output label (health_risk)
# ----------------------------------------------------------

def decode_output(pred):
    print("Decode the values")
    pred = [int(pred)]                     # ← Convert scalar → 1D array
    return encoders['health_risk'].inverse_transform(pred)[0]
    
 
# ----------------------------------------------------------
# 5. Predict function
# ----------------------------------------------------------

def make_prediction():
 
    # Get user input
    user_input = get_user_input() 
    print("Make Predictions")

    # Encode input
    encoded = encode_input(user_input)
   # encoded_x = encode_input(user_input)
    
   
    # Model prediction
    pred = load_model.predict(encoded)
   # pred = load_model.predict(encoded)[0]
    
    # Decode prediction
    decoded = decode_output(pred)
    # Decode output label
    #output = encoders["health_risk"].inverse_transform([pred])[0]


    print("\n====== Prediction Result ======")
    print("Encoded prediction:", int(pred[0]))
    print("Health Risk:", decoded)
    print("================================\n")



```
